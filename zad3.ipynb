3.3. Stosując kombinatoryczną interpretację liczb Stirlinga drugiego rodzaju i współczynników dwumianowych wykaż, że dla dowolnych liczb całkowitych $n\geq m\geq 0$ zachodzi równość:


$$
{n+1\brace m+1} = \sum^{n}_{k=1} {{n}\choose{k}} {k\brace m}
$$

Wywnioskuj z powyższej równości, że $B_{n+1}=\sum^{n}_{k=0} {{n}\choose{k}} B_k$

## Rozwiązanie
Rozważmy następujący model: mamy $n$ elementowy zbiór $X$ oraz $m$ elementowy podział tego zbioru, a następnie dokładamy kolejny element, oznaczmy go jako 0. Wtedy nowy element może być umieszczony w jednym z istniejących podziałów lub w nowym podziale. Przez co nowa ilość podziałów wyniesie $m+1$. To znaczy, że łączna liczba podziałów wyniesie:


$$
{m+1 \brace n+1}
$$

Z oryginalnego zbioru $X$ wybieramy $k$ elementów które następnie rozkładamy w $m$ podzbiorach, a w $m+1$-szym podziale umieszczamy pozostałe $n-k+1$ elementów. W taki sposób będziemy w stanie rozważyć przypadek, że nowy element znajduje się w $n-k+1$-elementowym podziale.
Dla ustalonego $k$ dostalibyśmy, że suma ilości takich podziałów jest równa (na podstawie zasady iloczynu):


$$
{{n}\choose{k}} {k\brace m}
$$

By uwzględnić wszystkie możliwe przypadki należałoby zsumować takie podziały dla wszystkich dopuszczalnych wartości $k$. Zauważmy, że najmniejsza liczba podziałów będzie równa 1, z czego wynika, że najmniejszą wartością $k$ jest 0. Z drugiej strony ponieważ liczby Stirlinga drugiego rodzaju jeśli $x > y$ to:

$$
{x \brace y} = 0
$$

Stąd widzimy, że maksymalną wartością $k$ jest $m$. 
Podsumowując, łączna liczba sumy wszystkich $m+1$ podziałów zbioru $n+1$-elementowego wynosi:

$$
{n+1\brace m+1} = \sum^{n}_{k=1} {{n}\choose{k}} {k\brace m}
$$

Otrzymaliśmy równość zadaną w poleceniu.
